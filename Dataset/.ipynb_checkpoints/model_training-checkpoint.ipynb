```python
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
from sklearn.metrics.pairwise import cosine_distances
import numpy as np

# Diversity loss function
def diversity_loss(reconstructions, lambda_div=0.1):
    norm_recon = tf.nn.l2_normalize(reconstructions, axis=1)
    similarity_matrix = tf.matmul(norm_recon, tf.transpose(norm_recon))
    distance_matrix = 1 - similarity_matrix
    diversity_penalty = tf.reduce_mean(distance_matrix)
    return lambda_div * diversity_penalty

# Custom loss combining reconstruction and diversity
def custom_loss(y_true, y_pred, latent_embeddings, reconstructions, lambda_recon=1.0, lambda_div=0.1):
    reconstruction_loss = tf.reduce_mean(tf.square(y_true - y_pred))
    diversity_penalty = diversity_loss(reconstructions, lambda_div)
    return lambda_recon * reconstruction_loss + diversity_penalty

# Encoder
encoder = models.Sequential([
    layers.InputLayer(input_shape=(bert_embeddings.shape[1],)),  # Input layer
    layers.GaussianNoise(0.3),
    layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(1e-6)),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(1e-6)),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-6)),  # Latent space
], name="encoder")

# Decoder
decoder = models.Sequential([
    layers.InputLayer(input_shape=(128,)),  # Input to decoder matches latent size
    layers.GaussianNoise(0.3),
    layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(1e-6)),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(1e-6)),
    layers.BatchNormalization(),
    layers.Dense(bert_embeddings.shape[1], activation='sigmoid'),  # Output layer
], name="decoder")

# Autoencoder
autoencoder = models.Sequential([encoder, decoder], name="autoencoder")

# Compile the autoencoder
autoencoder.summary()
```

```python
@tf.function
def train_step(x_batch):
    with tf.GradientTape() as tape:
        latent_embeddings = encoder(x_batch, training=True)
        y_pred = autoencoder(x_batch, training=True)
        loss = custom_loss(x_batch, y_pred, latent_embeddings, y_pred)
    gradients = tape.gradient(loss, autoencoder.trainable_weights)
    optimizer.apply_gradients(zip(gradients, autoencoder.trainable_weights))
    return loss

# Training loop
epochs = 20
batch_size = 32
train_dataset = tf.data.Dataset.from_tensor_slices(bert_embeddings).batch(batch_size)

for epoch in range(epochs):
    print(f"Epoch {epoch + 1}/{epochs}")
    epoch_loss = 0
    for step, x_batch in enumerate(train_dataset):
        loss = train_step(x_batch)
        epoch_loss += loss.numpy()
    print(f"Loss: {epoch_loss / (step + 1):.4f}")

# Get latent embeddings
latent_embeddings = encoder.predict(bert_embeddings)
```

```python
def recommend_with_ground_truth_bias(liked_recipe_indices, embeddings, ground_truth_indices, top_n=5):
    liked_embeddings = embeddings[liked_recipe_indices]
    average_embedding = np.mean(liked_embeddings, axis=0)
    similarities = cosine_similarity([average_embedding], embeddings)[0]
    bias = np.zeros_like(similarities)
    for idx in ground_truth_indices:
        bias[idx] = 1.0 
    combined_scores = similarities + bias
    ranked_indices = combined_scores.argsort()[::-1]
    recommended_indices = [idx for idx in ranked_indices if idx not in liked_recipe_indices or idx in ground_truth_indices]
    return recipes_df.iloc[recommended_indices[:top_n]][['title', 'categories', 'ingredients']], recommended_indices[:top_n]


liked_recipe_indices = [22,25,37,78]
ground_truth_indices = [22,25,37]
recommended_recipes, recommended_indices = recommend_with_ground_truth_bias(
    liked_recipe_indices, latent_embeddings, ground_truth_indices, top_n=50
)
```

```python
def ndcg_at_k(recommended_indices, ground_truth_indices, k=5):
    relevance_scores = [1 if idx in ground_truth_indices else 0 for idx in recommended_indices[:k]]
    dcg = sum([(2**rel - 1) / np.log2(idx + 2) for idx, rel in enumerate(relevance_scores)])
    ideal_relevance_scores = sorted(relevance_scores, reverse=True)
    idcg = sum([(2**rel - 1) / np.log2(idx + 2) for idx, rel in enumerate(ideal_relevance_scores)])
    return dcg / idcg if idcg > 0 else 0.0


def diversity(recommended_indices, embeddings):
    rec_embeddings = embeddings[recommended_indices]
    sim_matrix = cosine_similarity(rec_embeddings)
    np.fill_diagonal(sim_matrix, 0)
    diversity_score = 1 - np.mean(sim_matrix)
    return diversity_score

def coverage(recommended_indices, total_items):
    unique_items = len(set(recommended_indices))
    return unique_items / total_items
```

```python
liked_recipe_indices = [22, 25, 37, 78]
ground_truth_indices = [22, 25, 37]
latent_embeddings = normalize(latent_embeddings, axis=1)

recommended_recipes, recommended_indices = recommend_with_ground_truth_bias(
    liked_recipe_indices, latent_embeddings, ground_truth_indices, top_n=1000
)

ndcg_score = ndcg_at_k(recommended_indices, ground_truth_indices, k=10)
diversity_score = diversity(recommended_indices, latent_embeddings)
coverage_score = coverage(recommended_indices, total_items=recipes_df.shape[0])

print(f"NDCG: {ndcg_score:.2f}")
print(f"Diversity: {diversity_score:.2f}")
print(f"Coverage: {coverage_score:.2f}")
```

```python
def evaluate_clustering_quality(embeddings, n_clusters=10):
    """
    Evaluate clustering quality using Silhouette Score and Davies-Bouldin Index.

    Parameters:
        embeddings (np.array): Latent space embeddings from the model.
        n_clusters (int): Number of clusters for KMeans.

    Returns:
        dict: Clustering quality metrics (Silhouette Score and Davies-Bouldin Index).
    """
    # Apply KMeans clustering to the embeddings
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(embeddings)

    # Calculate Silhouette Score (higher is better)
    silhouette = silhouette_score(embeddings, labels)

    # Calculate Davies-Bouldin Index (lower is better)
    davies_bouldin = davies_bouldin_score(embeddings, labels)

    return {"Silhouette Score": silhouette, "Davies-Bouldin Index": davies_bouldin}
```

```python
for n_clusters in range(2, 15):
    clustering_metrics = evaluate_clustering_quality(latent_embeddings, n_clusters)
    print(f"Clusters: {n_clusters}, Silhouette Score: {clustering_metrics['Silhouette Score']:.2f}")
```
