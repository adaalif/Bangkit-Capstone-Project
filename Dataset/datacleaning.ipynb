# insights from EDA

- Dari hasil analisis data, terdapat banyak nilai kosong yang perlu dibersihkan.
- Terdapat nilai kalori yang sangat tinggi hingga lebih dari 30.000, yang tidak realistis untuk satu porsi makanan. Untuk memastikan data lebih representatif, kita perlu menetapkan batasan (threshold) yang realistis. Kalori yang wajar untuk makanan bisa dibatasi, misalnya, antara 1.000 hingga 1.500 kalori untuk makanan berkalori tinggi.

- Kandungan sodium juga menunjukkan banyak outlier, dengan beberapa nilai melebihi batas konsumsi harian yang direkomendasikan. Mengingat rata-rata sodium untuk makanan berada di kisaran 800-1.000 mg, kita bisa menetapkan ambang batas sekitar 1.500 mg untuk menjaga data tetap relevan.

- Terdapat sejumlah data duplikat dalam dataset, yang perlu dihapus untuk mencegah redundansi dan kesalahan dalam analisis. Mengidentifikasi duplikat berdasarkan atribut penting seperti judul, kalori, dan bahan akan membantu kita mempertahankan data yang unik.

- Awalnya, data dengan nilai kalori di bawah 50 tampak seperti outlier. Namun, setelah ditinjau lebih lanjut, ternyata data ini merujuk pada resep makanan pendamping seperti saus atau kaldu.

# Data Cleaning

```python
data_withsteps.dropna(inplace=True)
```

```python
data_withsteps.isnull().sum()
```

```python
data_withsteps.describe()
```

```python
data_withsteps.info()
```

```python
data_withsteps.drop_duplicates(subset=['title', 'calories', 'protein', 'fat', 'sodium'], inplace=True)
```

```python
data_withsteps.info()
```

```python
data_withsteps.describe()
```

```python
data_withsteps = data_withsteps[data_withsteps['calories'] < 1500]
```

```python
data_withsteps.describe()
```

```python
data_withsteps = data_withsteps[data_withsteps['sodium'] < 1500]
```

```python
data_withsteps.describe()
```

```python
data_withsteps['total_steps'] = data_withsteps['directions'].apply(
    lambda x: len(x) if isinstance(x, list) else len(eval(x)) if isinstance(x, str) and x.startswith('[') else 0
)

data_withsteps['total_ingredients'] = data_withsteps['ingredients'].apply(
    lambda x: len(x) if isinstance(x, list) else len(eval(x)) if isinstance(x, str) and x.startswith('[') else 0
)
data_withsteps.head()
```

```python
print(data_withsteps[['total_steps', 'total_ingredients']].describe())
```

```python
data_withsteps[data_withsteps['total_steps'] == 2]
```

```python
data_withsteps[data_withsteps['total_ingredients'] < 3]
```

```python
data_withsteps = data_withsteps[data_withsteps['total_steps'] >= 2]
```
